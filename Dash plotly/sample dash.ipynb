```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import numpy as np
import plotly.graph_objs as go
```

```python
app = dash.Dash()
```

```python
colors = {
    'text': '#ff0000',
    'plot_color': '#D3D3D3',
    'paper_color': '#00FFFF'
}
```

```python
app.layout = html.Div([
    html.H1(children='Hello Dash!!!',
            style = {
                'textAlign': 'center',
                'color': colors['text']
                
            }
             ),
    html.Div(children= 'Dash of a data product ',
             style = {
                'textAlign': 'center',
                'color': colors['text'],
                
            }
             ),
    dcc.Graph(
         id = 'Samplechart',
        figure = {
            'data': [
                {'x' : [1,2,3], 'y':[5,10,15],'type':'bar','name':'First chart'},
                {'x' : [1,2,3], 'y':[20,25,35],'type':'bar','name':'second chart'}
            ],
            'layout': {
                       'plot_bgcolor' :  colors['plot_color'],
                       'paper_bgcolor' :  colors['paper_color'],
                       'font' : {
                           'color': colors['text']
                       },
                        'title': 'Simple Bar chart'
                }           
        }
    )
])
```

```python
if __name__ =='__main__': 
        app.run(port=8060)
```

```python
np.random.seed(50)
x_rand = np.random.randint(1,61,60)
y_rand = np.random.randint(1,61,60)
```

```python
app.layout = html.Div([
    dcc.graph(
        id='scatter_chart',
        figure = {
            'data': {
              go.scatter{
                  x= x_rand,
                   y= y_rand,
                   mode = 'marker'
              } 
            },
            'layout': go.layout(
                title = 'scatterplot of Random 60 points',
                xaxis= {'title': 'Random X values'},
                yaxis= {'title':  'Random Y values'}
            )
        }
    )
])
```

```python
if __name__ =='__main__': 
        app.run(port=8060)
```
